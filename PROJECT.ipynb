# TEXT PRE PROCESSING

```python
data['text'] = data['text'].str.lower().replace(r'^\w\s',' ').replace(r'\n',' ',regex = True)
```

```python
import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
from sklearn.feature_extraction.text import TfidfVectorizer as tfidf
from sklearn.metrics.pairwise import cosine_similarity as cose
```

```python
nltk.download('punkt_tab')
```

```python
stemming = PorterStemmer()
def tokenize_words(text):
  token_words = word_tokenize(text)
  stemmed_tokens = [stemming.stem(i) for i in token_words]
  return ' '.join(stemmed_tokens)
```

```python
data['text'].apply(lambda i: tokenize_words(i))
```

```python
tf = tfidf(analyzer='word',stop_words='english')  # analyzer='word': This parameter within the tfid function specifies that the analysis should be performed on individual words within the text. By default, TfidfVectorizer considers individual words for analysis, and this parameter confirms that setting.
```

```python
vectors = tf.fit_transform(data['text'])
similar = cose(vectors) #Similar songs
```

# RECCOMENDING

```python
def reccommend_song(song):
  song_index = data[data['song'] == song].index[0]
  recom_distance = sorted(list(enumerate(similar[song_index])),reverse=True,key=lambda x:x[1])
  songs = []
  for i in recom_distance[1:6]:
    songs.append(data.iloc[i[0]].song)
  return songs
```

```python
ans = reccommend_song('Someday At Christmas')
ans
```
